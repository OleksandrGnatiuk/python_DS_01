# 1. Створіть одновимірний масив (вектор) з першими 10-ма натуральними числами та виведіть його значення.

```python
import numpy as np

# Варіант 1: з використанням стандартних методів python, наприклад 'range', але всі ці методи працюють повільніше за методи бібліотеки numpy
v = np.array(range(1,11))
v
```

```python
# варіант 2: за допомогою функції 'arange'. Особливість - дозволяє встановити крок, в т.ч. не кратний 1:
v1 = np.arange(1,11, 1)
v1
```

```python
# варіант 3: за допомогою функції 'linspace' (дозволяє встановите стартове та кінцеве значення, а також кількіть даних який необхідний в цьому проміжку):
v_1 = np.linspace(1,10, num=10, dtype=int)
v_1
```

# 16. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
